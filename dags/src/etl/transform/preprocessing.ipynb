### 1. Dữ liệu giá lịch sử chứng khoán của VN30

```python
import os
import pandas as pd

# Đường dẫn đến thư mục chứa các file CSV
folder_path = r'D:\project_persional\chabot_stock\data\raw\quote'

# Lấy danh sách tất cả các file CSV trong thư mục
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Tạo một danh sách để lưu trữ các bảng dữ liệu
dfs = []

# Đọc từng file CSV và thêm vào danh sách
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    dfs.append(df)

# Hợp nhất tất cả các bảng dữ liệu thành một
quote_summary = pd.concat(dfs, ignore_index=True)

# Đường dẫn để lưu file CSV mới
output_path = r'D:\project_persional\chabot_stock\data\raw\Quote_summary.csv'

# Lưu bảng dữ liệu hợp nhất vào file CSV
quote_summary.to_csv(output_path, index=False, encoding='utf-8-sig')

print(f"Tất cả các file CSV đã được hợp nhất và lưu vào {output_path}")
```

### 2. Tổng hợp dữ liệu báo cáo tài chính

```python
import os
import pandas as pd

# Đường dẫn đến thư mục chứa các file CSV
folder_path = r'D:\project_persional\chabot_stock\data\processed\finance'

# Danh sách các loại thông tin cần phân loại
info_types = ['ratio']

# Tạo từ điển để lưu danh sách các file theo loại thông tin
file_dict = {info: [] for info in info_types}

# Lấy danh sách tất cả các file CSV trong thư mục
csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

# Phân loại các file theo loại thông tin dựa trên tên file
for file in csv_files:
    for info in info_types:
        if info in file:
            file_dict[info].append(file)
            break

# Hàm để hợp nhất các file CSV theo loại thông tin
def merge_files(files, output_file):
    dfs = []
    for file in files:
        file_path = os.path.join(folder_path, file)
        df = pd.read_csv(file_path)
        dfs.append(df)
    if dfs:  # Kiểm tra nếu có dữ liệu để hợp nhất
        summary_df = pd.concat(dfs, ignore_index=True)
        summary_df.to_csv(output_file, index=False, encoding='utf-8-sig')
        print(f"Tất cả các file {info} đã được hợp nhất và lưu vào {output_file}")
    else:
        print(f"Không có file nào cho {info}")

# Đường dẫn để lưu các file CSV hợp nhất
output_paths = {
    'ratio': r'D:\project_persional\chabot_stock\data\Ratio_summary.csv'
}

# Hợp nhất và lưu các file theo từng loại thông tin
for info in info_types:
    files = file_dict[info]
    output_file = output_paths[info]
    merge_files(files, output_file)
```

# Mô tả dữ liệu

```python
import pandas as pd

def describe_data(file_path):
    """
    Hàm mô tả dữ liệu từ file CSV: hiển thị thông tin cột, dữ liệu mẫu, thống kê cơ bản và giá trị thiếu.
    Input: file_path (đường dẫn đến file CSV)
    Output: In ra thông tin mô tả dữ liệu
    """
    print(f"--- Mô tả dữ liệu cho file: {file_path} ---")
    
    # Đọc file CSV
    df = pd.read_csv(file_path)
    
    # Hiển thị thông tin về các cột (tên cột, kiểu dữ liệu, số lượng giá trị không null)
    print("\nThông tin các cột:")
    print(df.info())
    
    # Hiển thị 5 dòng đầu tiên của dữ liệu
    print("\n5 dòng đầu tiên:")
    print(df.head())
    
    # Thống kê cơ bản cho các cột số (min, max, mean, v.v.)
    print("\nThống kê cơ bản:")
    print(df.describe())
    
    # Kiểm tra số lượng giá trị thiếu trong mỗi cột
    print("\nSố lượng giá trị thiếu trong mỗi cột:")
    print(df.isnull().sum())
```

```python
import pandas as pd
import os

# Đường dẫn thư mục nguồn và đích
source_folder = r'D:\project_persional\chabot_stock\data\raw\finance'
output_folder = r'D:\project_persional\chabot_stock\data\processed'

# Tạo thư mục output nếu chưa tồn tại
os.makedirs(output_folder, exist_ok=True)

desired_columns = [
    'CP', 'Năm', 'Kỳ', 'Nợ/VCSH', 'TSCĐ / Vốn CSH', 'Vốn CSH/Vốn điều lệ',
    'Biên lợi nhuận ròng (%)', 'ROE (%)', 'ROIC (%)', 'ROA (%)', 'Tỷ suất cổ tức (%)',
    'Đòn bẩy tài chính', 'Vốn hóa (Tỷ đồng)', 'Số CP lưu hành (Triệu CP)',
    'P/E', 'P/B', 'P/S', 'P/Cash Flow', 'EPS (VND)', 'BVPS (VND)',
    '(Vay NH+DH)/VCSH', 'Vòng quay tài sản', 'Vòng quay TSCĐ',
    'Số ngày thu tiền bình quân', 'Số ngày tồn kho bình quân',
    'Số ngày thanh toán bình quân', 'Chu kỳ tiền', 'Vòng quay hàng tồn kho',
    'Biên EBIT (%)', 'Biên lợi nhuận gộp (%)', 'EBITDA (Tỷ đồng)', 'EBIT (Tỷ đồng)',
    'Chỉ số thanh toán hiện thời', 'Chỉ số thanh toán tiền mặt',
    'Chỉ số thanh toán nhanh', 'Khả năng chi trả lãi vay', 'EV/EBITDA'
]

# Lấy danh sách file ratio
ratio_files = [f for f in os.listdir(source_folder) 
               if f.endswith('.csv') and 'ratio' in f.lower()]

merged_dfs = []  # List để lưu các DataFrame đã xử lý

# Xử lý từng file
for file_name in ratio_files:
    source_path = os.path.join(source_folder, file_name)
    
    # Đọc file CSV, không có header
    df = pd.read_csv(source_path, header=None)
    
    # 2. Lấy dòng thứ hai (index 1) làm tiêu đề cột
    new_columns = df.iloc[1]
    df = df[2:]  # Bỏ hai dòng đầu tiên
    
    # 3. Đặt lại tiêu đề cột
    df.columns = new_columns
    
    # 5. Đảm bảo tất cả các cột mong muốn có mặt, thêm cột thiếu với giá trị Null
    for col in desired_columns:
        if col not in df.columns:
            df[col] = pd.NA  # Thêm cột với giá trị Null
    
    # 6. Sắp xếp lại các cột theo thứ tự mong muốn
    df = df[desired_columns]
    
    # Reset index
    df = df.reset_index(drop=True)
    
    merged_dfs.append(df)  # Thêm DataFrame đã xử lý vào list

# Hợp nhất và lưu file tổng hợp
if merged_dfs:
    summary_df = pd.concat(merged_dfs, ignore_index=True)
    output_summary_path = os.path.join(output_folder, 'process_ratio.csv')
    summary_df.to_csv(output_summary_path, index=False, encoding='utf-8-sig')
    print(f"Tất cả các file ratio đã được hợp nhất và lưu vào: {output_summary_path}")
else:
    print("Không có file nào để xử lý")
```

### 1.1 Hàm xử lý lịch sử giá

```python
import os
import pandas as pd

def process_and_save_quote_data(input_folder, output_file):
    """
    Hàm xử lý và lưu lịch sử giá từ các file CSV trong một thư mục.
    
    Parameters:
    - input_folder (str): Đường dẫn đến thư mục chứa các file CSV.
    - output_file (str): Đường dẫn đến file CSV đầu ra (process_quote.csv).
    
    Returns:
    - None: Kết quả được lưu trực tiếp vào file CSV.
    """
    # Lấy danh sách tất cả các file CSV trong thư mục
    csv_files = [f for f in os.listdir(input_folder) if f.endswith('.csv')]
    
    # Kiểm tra xem có file CSV nào trong thư mục không
    if not csv_files:
        print("Không có file CSV nào trong thư mục.")
        return
    
    # Tạo danh sách để lưu trữ các bảng dữ liệu
    dfs = []
    
    # Đọc từng file CSV và thêm vào danh sách
    for file in csv_files:
        file_path = os.path.join(input_folder, file)
        df = pd.read_csv(file_path)
        dfs.append(df)
    
    # Hợp nhất tất cả các bảng dữ liệu thành một
    quote_summary = pd.concat(dfs, ignore_index=True)
    
    # Chuyển đổi cột 'volume' sang kiểu float64
    quote_summary['volume'] = quote_summary['volume'].astype('float64')
    
    # Lưu bảng dữ liệu đã xử lý vào file CSV
    quote_summary.to_csv(output_file, index=False, encoding='utf-8-sig')
    
    print(f"Đã xử lý và lưu kết quả vào {output_file}")

# Ví dụ sử dụng hàm
input_folder = r'D:\project_persional\chabot_stock\data\raw\quote'
output_file = r'D:\project_persional\chabot_stock\data\processed\process_quote.csv'
process_and_save_quote_data(input_folder, output_file)
```

### 1.3 Hàm xử lý Chỉ số tài chính

```python
import pandas as pd
import os

def process_and_save_ratio_data(source_folder, output_folder):
    """
    Hàm xử lý và lưu các file chỉ số tài chính (ratio) từ thư mục nguồn vào một file CSV hợp nhất trong thư mục đích.
    
    Parameters:
    - source_folder (str): Đường dẫn đến thư mục chứa các file CSV nguồn.
    - output_folder (str): Đường dẫn đến thư mục để lưu file CSV hợp nhất.
    
    Returns:
    - None: Kết quả được lưu trực tiếp vào file CSV.
    """
    # Tạo thư mục output nếu chưa tồn tại
    os.makedirs(output_folder, exist_ok=True)
    
    # Danh sách các cột mong muốn
    desired_columns = [
        'CP', 'Năm', 'Kỳ', 'Nợ/VCSH', 'TSCĐ / Vốn CSH', 'Vốn CSH/Vốn điều lệ',
        'Biên lợi nhuận ròng (%)', 'ROE (%)', 'ROIC (%)', 'ROA (%)', 'Tỷ suất cổ tức (%)',
        'Đòn bẩy tài chính', 'Vốn hóa (Tỷ đồng)', 'Số CP lưu hành (Triệu CP)',
        'P/E', 'P/B', 'P/S', 'P/Cash Flow', 'EPS (VND)', 'BVPS (VND)',
        '(Vay NH+DH)/VCSH', 'Vòng quay tài sản', 'Vòng quay TSCĐ',
        'Số ngày thu tiền bình quân', 'Số ngày tồn kho bình quân',
        'Số ngày thanh toán bình quân', 'Chu kỳ tiền', 'Vòng quay hàng tồn kho',
        'Biên EBIT (%)', 'Biên lợi nhuận gộp (%)', 'EBITDA (Tỷ đồng)', 'EBIT (Tỷ đồng)',
        'Chỉ số thanh toán hiện thời', 'Chỉ số thanh toán tiền mặt',
        'Chỉ số thanh toán nhanh', 'Khả năng chi trả lãi vay', 'EV/EBITDA'
    ]
    
    # Lấy danh sách file ratio
    ratio_files = [f for f in os.listdir(source_folder) 
                   if f.endswith('.csv') and 'ratio' in f.lower()]
    
    # Kiểm tra xem có file nào để xử lý không
    if not ratio_files:
        print("Không có file ratio nào để xử lý.")
        return
    
    merged_dfs = []  # List để lưu các DataFrame đã xử lý
    
    # Xử lý từng file
    for file_name in ratio_files:
        source_path = os.path.join(source_folder, file_name)
        
        # Đọc file CSV, không có header
        df = pd.read_csv(source_path, header=None)
        
        # Lấy dòng thứ hai (index 1) làm tiêu đề cột
        new_columns = df.iloc[1]
        df = df[2:]  # Bỏ hai dòng đầu tiên
        
        # Đặt lại tiêu đề cột
        df.columns = new_columns
        
        # Đảm bảo tất cả các cột mong muốn có mặt, thêm cột thiếu với giá trị Null
        for col in desired_columns:
            if col not in df.columns:
                df[col] = pd.NA  # Thêm cột với giá trị Null
        
        # Sắp xếp lại các cột theo thứ tự mong muốn
        df = df[desired_columns]
        
        # Reset index
        df = df.reset_index(drop=True)
        
        merged_dfs.append(df)  # Thêm DataFrame đã xử lý vào list
    
    # Hợp nhất và lưu file tổng hợp
    if merged_dfs:
        summary_df = pd.concat(merged_dfs, ignore_index=True)
        output_summary_path = os.path.join(output_folder, 'process_ratio.csv')
        summary_df.to_csv(output_summary_path, index=False, encoding='utf-8-sig')
        print(f"Tất cả các file ratio đã được hợp nhất và lưu vào: {output_summary_path}")
    else:
        print("Không có dữ liệu để hợp nhất.")


source_folder = r'D:\project_persional\chabot_stock\data\raw\finance'
output_folder = r'D:\project_persional\chabot_stock\data\processed'
process_and_save_ratio_data(source_folder, output_folder)
```

### 1.4 Hàm xử lý thông tin công ty

```python
import os
import pandas as pd

def process_and_save_company_data(source_folder, output_folder):
    """
    Hàm xử lý và lưu thông tin công ty từ các file CSV trong thư mục nguồn vào một file CSV hợp nhất.
    
    Parameters:
    - source_folder (str): Đường dẫn đến thư mục chứa các file CSV nguồn.
    - output_folder (str): Đường dẫn đến thư mục để lưu file CSV hợp nhất.
    
    Returns:
    - None: Kết quả được lưu trực tiếp vào file CSV.
    """
    # Danh sách các loại thông tin cần phân loại
    info_types = ['overview', 'profile']
    
    # Tạo từ điển để lưu danh sách các file theo loại thông tin
    file_dict = {info: [] for info in info_types}
    
    # Lấy danh sách tất cả các file CSV trong thư mục
    csv_files = [f for f in os.listdir(source_folder) if f.endswith('.csv')]
    
    # Kiểm tra xem có file CSV nào không
    if not csv_files:
        print("Không có file CSV nào trong thư mục nguồn.")
        return
    
    # Phân loại các file theo loại thông tin dựa trên tên file
    for file in csv_files:
        for info in info_types:
            if info in file:
                file_dict[info].append(file)
                break
    
    # Hàm con để hợp nhất các file CSV theo loại thông tin
    def merge_files(files, output_file):
        dfs = []
        for file in files:
            file_path = os.path.join(source_folder, file)
            df = pd.read_csv(file_path)
            dfs.append(df)
        if dfs:  # Kiểm tra nếu có dữ liệu để hợp nhất
            summary_df = pd.concat(dfs, ignore_index=True)
            summary_df.to_csv(output_file, index=False, encoding='utf-8-sig')
            return summary_df
        else:
            print(f"Không có file nào cho {info}")
            return None
    
    # Đường dẫn tạm thời để lưu các file CSV hợp nhất
    temp_output_paths = {
        'overview': os.path.join(source_folder, 'Overview_summary.csv'),
        'profile': os.path.join(source_folder, 'Profile_summary.csv')
    }
    
    # Hợp nhất các file theo từng loại thông tin
    profile_df = None
    overview_df = None
    for info in info_types:
        files = file_dict[info]
        output_file = temp_output_paths[info]
        if info == 'profile':
            profile_df = merge_files(files, output_file)
        elif info == 'overview':
            overview_df = merge_files(files, output_file)
    
    # Kiểm tra xem có dữ liệu để xử lý tiếp không
    if profile_df is None or overview_df is None:
        print("Không thể xử lý tiếp do thiếu dữ liệu từ profile hoặc overview.")
        return
    
    # Xử lý dữ liệu: Loại bỏ các cột không cần thiết
    profile_df = profile_df.drop(columns=['history_dev', 'company_promise', 'key_developments', 
                                          'MaCK', 'business_risk', 'company_profile', 
                                          'business_strategies'], errors='ignore')
    
    overview_df = overview_df.drop(columns=['exchange', 'no_shareholders', 'foreign_percent',
                                            'outstanding_share', 'issue_share', 'established_year',
                                            'no_employees', 'delta_in_week', 'delta_in_month',
                                            'delta_in_year', 'website', 'industry_id', 
                                            'industry_id_v2', 'MaCK'], errors='ignore')
    
    # Gộp hai DataFrame theo cột 'symbol'
    company_summary = pd.merge(profile_df, overview_df, on='symbol', how='inner')
    
    # Tạo thư mục đích nếu chưa tồn tại
    os.makedirs(output_folder, exist_ok=True)
    
    # Lưu kết quả cuối cùng
    output_file = os.path.join(output_folder, 'process_company.csv')
    company_summary.to_csv(output_file, index=False, encoding='utf-8-sig')
    print(f"Đã lưu file tổng hợp vào: {output_file}")


source_folder = r'D:\project_persional\chabot_stock\data\raw\company'
output_folder = r'D:\project_persional\chabot_stock\data\processed'
process_and_save_company_data(source_folder, output_folder)
```

```python

import pandas as pd
def processing_dimtime(file_path):
    df_quote = pd.read_csv(file_path)
    # Chuyển đổi cột 'time' sang định dạng datetime
    df_quote['time'] = pd.to_datetime(df_quote['time'], format='%Y-%m-%d')
    # Giả sử df_quote đã được định nghĩa và cột 'time' đã chuyển sang datetime
    # Tách các thành phần thời gian
    df_quote['Date'] = df_quote['time']
    df_quote['Year'] = df_quote['time'].dt.year
    df_quote['Quarter'] = df_quote['time'].dt.quarter
    df_quote['Month'] = df_quote['time'].dt.month
    df_quote['Week'] = df_quote['time'].dt.isocalendar().week  # Sử dụng ISO week
    df_quote['Day'] = df_quote['time'].dt.day
    
    # Tạo TimeKey từ Date (định dạng YYYYMMDD)
    df_quote['TimeKey'] = df_quote['time'].dt.strftime('%Y%m%d').astype(int)
    
    # Tạo DataFrame Dim_Time và loại bỏ bản ghi trùng lặp
    dim_time_df = df_quote[[
        'TimeKey', 'Date', 'Year', 'Quarter', 'Month', 'Week', 'Day'
    ]].drop_duplicates()
    return dim_time_df
```

```python
import pandas as pd
from sqlalchemy import insert
from sqlalchemy.ext.asyncio import AsyncSession
from config.database import get_db  # Import từ file config của bạn

async def insert_df_to_table(model, df: pd.DataFrame):
    """
    Hàm chèn dữ liệu từ DataFrame vào bảng trong database bất đồng bộ.
    
    Args:
        model: SQLAlchemy model class (ví dụ: DimTime)
        df: Pandas DataFrame chứa dữ liệu cần chèn
    """
    async with get_db() as session:  # Sử dụng get_db() từ config của bạn
        try:
            # Lấy tên cột primary key từ model
            pk_col = model.__table__.primary_key.columns[0].name

            # Lấy danh sách các primary key đã tồn tại
            existing_pks = {x[0] for x in await session.execute(
                f"SELECT {pk_col} FROM {model.__tablename__}"
            )}

            # Lọc các bản ghi chưa tồn tại
            new_df = df[~df[pk_col].isin(existing_pks)]

            if new_df.empty:
                print("Không có bản ghi mới cần thêm.")
                return

            # Chuyển DataFrame thành list dictionaries
            data = new_df.to_dict(orient='records')

            # Thực hiện chèn dữ liệu
            await session.execute(
                insert(model).values(data)
            )
            print(f"Đã thêm {len(data)} bản ghi vào {model.__tablename__}")

        except Exception as e:
            print(f"Lỗi khi chèn dữ liệu: {str(e)}")
            await session.rollback()
            raise e

# Ví dụ sử dụng với DimTime
# await insert_df_to_table(DimTime, dim_time_df)
```
